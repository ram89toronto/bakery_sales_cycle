# Bakery Data - Analysis: Sales Cycle of Products


## Importing Libraries

```python
import datetime as dt
import pandas as pd
import numpy as np
import calendar as cl
from matplotlib import pyplot as plt
%matplotlib inline
```

## Loading Bakery Data into Pandas DataFrame

```python
data= pd.read_csv("BakeryData.csv")
```

## Inspection of loaded data

```python
print(data.head())
print(data.tail())
print(data.info())
print(data.describe())
```

# Conversion coloumns into computable datatypes

```python
def date_conversion(x):
    a=dt.datetime.strptime(x,"%m/%d/%Y")
    return a
def date_months(x):
    a = int(x["dates"].month)
    return a
def date_days(x):
    a = int(x["dates"].day)
    return a
def date_weekdays(x):
    a = x["dates"].weekday()
    return str(cl.day_name[a])

def line_plot(data_set,splot):
    splot.plot(data_set.index,data_set.Cakes)
    splot.plot(data_set.index,data_set.Pies)
    splot.plot(data_set.index,data_set.Cookies)
    splot.plot(data_set.index,data_set.Smoothies)
    splot.plot(data_set.index,data_set.Coffee)
    splot.legend(labels=data_set.columns[0:5])
    #splot.title(" Sales Cycle ")
    splot.set_ylabel(" Average Sales quantity")
    return 
```

```python
data["dates"] = data["Date"].apply(lambda row: date_conversion(row))
```

```python
data.head()
```

```python
data = data.drop("Date",1)
```

```python
data.head()
```

```python
data["month"]= data.apply(lambda row: date_months(row),axis=1)
data["day"]= data.apply(lambda row: date_days(row),axis=1)
data["weekday"]= data.apply(lambda row: date_weekdays(row),axis=1)

data.tail()
```

```python
data_weekday = data.groupby(["weekday"]).mean()
data_month = data.groupby(["month"]).mean()
data_day = data.groupby(["day"]).mean()

# Checking the cleaned Data
data_day.head()
```

```python
data_month.head()
```

## Ploting the sales cycle by week , days and month .



```python


fig= plt.figure(figsize=(20,20),dpi=200)
ax= plt.subplot(3,1,1)
line_plot(data_weekday,ax)
ax.set_title("BAKERY SALES CYCLE - WEEK wise")
bx =plt.subplot(3,1,3)
line_plot(data_month,bx)
bx.set_title("BAKERY SALES CYCLE - Month wise")
cx= plt.subplot(3,1,2)
line_plot(data_day,cx)
cx.set_title("BAKERY SALES CYCLE - Day wise")
plt.show()

```

## Ploting sales cycle gives insight to which day of week, which  month of the year for a bakery owner to optimize his production and maximize sales using right promotional offer over last year.
